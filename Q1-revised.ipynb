```python
you=0
computer=0
import random

input0 = int(input("How many games you count as a win:\n"))

print("You can type quit if you want to exit\n")

while you < input0 and computer < input0:
    input1=input("rock, scissors, paper????:\n").lower()
    
    if input1  == "quit": 
        break
    else: 
        pass
    
    a = random.randint(1, 3)
    if a==1:
        b = "rock"
    elif a==2:
        b="scissors"
    elif "a==3":
        b="paper"
    else:
        print("please type rock,scissors or paper\n")

    print(f"You chose {input1} \t  Computer randomed {b}\n")

    if input1 == b :
        print("This Round: Tie\n");
    elif input1 == "rock" and a == 2:
        print("This Round: You Won\n")
        you +=1
    elif input1 == "rock" and a == 3:
        print("This Round: You Lost\n")
        computer +=1
    elif input1 == "scissors" and a == 3:
        print("This Round: You Won\n")
        you +=1
    elif input1 == "scissors" and a == 1:
        print("This Round: You Lost\n")
        computer +=1
    elif input1 == "paper" and a == 1:
        print("This Round: You Won\n")
        you +=1
    elif input1 == "paper" and a == 2:
        print("This Round: You Lost\n")
        computer +=1
    else:
        print("Please type one of the following: rock,scissors or paper\n")


    if computer == input0:
        print("OVERALL :COMPUTER VICTORY")
    elif you==input0:
        print("OVERALL: You somehow managed to win.")
    print(f"You:{you} \t Computer: {computer}")
    
```
