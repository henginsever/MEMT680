```python
import random 
class Dice(): #Class for dice roll and keeping  dices as a self name.
    def __init__(self): 
        self.dice1 = 0
        self.dice2 = 0
        self.total = 0 # everything is 0 at the moment, just for error handling purposes.
    
    def roll(self): 
        self.dice1 = random.randint(1, 6) #dice 1 
        self.dice2 = random.randint(1, 6) #dice 2
        self.total = self.dice1 + self.dice2 #essentially, all i need is sum of two dices so i dont need to use both dice values again and again.
        print(f"Dices are rolled: {self.dice1} and {self.dice2}") #printing incoming dices
        return self.dice1, self.dice2, self.total 
```

```python
class Table(Dice):
    def __init__(self):
        super().__init__() #inheriting every attributes defined at the parent class.
        
    def point_check(self):
        super().roll()
        #print(f"Dices are {self.dice1} and {self.dice1}")
        self.point = self.dice1+self.dice2
        if 0 < self.point < 13: 
            return print("point has been set")
        else: 
            print("point has not been set")  
```

```python
class Player(Table):
    def __init__(self):
        super().__init__() #inheriting every attributes defined at the parent classes.
        self.name = input("Please enter your name:") #calling name of the player
        
        while True:
            try:
                self.bankroll = input("Please enter your bankroll:")
                if int(self.bankroll)<0: #checking for negative inputs
                    print("you can not enter negative number")
                    continue #continue is used for sending user to the start of the loop.
                else:
                    value = int(self.bankroll) #this is the value that is considered acceptable
                    print("you good boi")
            except:
                try:
                    value = float(self.bankroll) #even though the number is positive, only integer is allowed.
                    print("Please enter the value as an integer")
                    continue
                except ValueError: #anything other than number is not allowed
                    print("That's not a number!")
                    continue
            else:
                self.bankroll = int(self.bankroll) #to be sure that self.bankroll is saved as an integer number.
                return False  #i returned "False" to end loop.
```

```python
class Bets(Player):
    def __init__(self):
        super().__init__() #inheriting every attributes defined at the parent classes.
        self.point_on = 0  
        self.pas = 0
    
    def Insufficient_funds(self):
        while True:
            try:
                self.bet_amount = int(input(f"Your bankroll is equal to {self.bankroll}. Please enter an amount less than your bankroll:")) #asking for a re-entry of bet_amount
                if int(self.bet_amount) > int(self.bankroll): #after re-entry checking if it exceeds bankroll or not.
                    continue
                elif self.bet_amount < 0: #checking for negative number detection
                    print("You can not enter negative bet amount") 
                    continue
                else: 
                    break #sometimes i used break instead of returning False. Sometimes i encountered errors such as endless loop so i tend to keep it if is working 
            except: 
                continue
            else: 
                break
    
    def pass_line(self): #pass line code 
        self.bankroll -= self.bet_amount #updating bankroll
        if self.total == 7 or self.total == 11: #win condition for pass line
            self.output = self.bet_amount 
            Bets.bet_winner(self) 
        elif self.total == 2 or self.total == 3: #lose condition for pass line
            Bets.bet_loser(self)
        elif self.total == 0: #if dices were not rolled until this method is called, it is roleld now
            Table.point_check(self)
            self.bankroll += self.bet_amount #since we will go back to the beginning of function, i might double cut the bet_amount. That is why i am returning same value for such occasions
            Bets.pass_line(self)
        else: 
            print("Ladies and gentlemen! we got on") #come out roll phase did not concluded
            self.point_on = self.total #point is assigned for next rolls
            self.pas = 1 #this is for shooter phase. Can be considered as an identification that the user chose pass_line
    
    def do_not_pass(self): #basically same thing like the previous method pass_line
        self.bankroll -= self.bet_amount
        #self.betsave = self.bet_amount 
        if self.total == 2 or self.total == 3:
            Bets.bet_winner(self) 
        elif self.total == 7 or self.total == 11:
            Bets.bet_loser(self)
        elif self.total == 0:
            Table.point_check(self)
            self.bankroll += self.bet_amount
            Bets.do_not_pass(self)
        else: 
            print("Ladies and gentlemen! we got on")
            self.point_on = self.total
            self.pas = 2
    
    def betting_turn(self):
        self.bet_amount = 0 #this was adjustment for error handling purposes. However since i have defined refresh method, this might not be necessary at all. But i tend to keep it because the code is working fine.
        answer_for_bet = input("Do you wish to bet this round y/n?:").lower()
        
        if answer_for_bet == "n":
            return print(f"{self.name} passed this round")
        
        elif answer_for_bet == "y":
            Bets.ingest_bet(self) #i call the values first because i wrote all codes to utilize bet_amount(checking conditions require pre-defined bet_amount
            
            while True: #only allows to input to be used.
                self.bet_type = input("for bet type, enter one of the following: pass or don't pass:").lower()
                try:
                    if (self.bet_type).lower() == "pass":
                        Bets.pass_line(self)    
                    elif (self.bet_type).lower() == "don't pass":
                        Bets.do_not_pass(self)
                    else:
                        print("please type one of the following pass, don't pass or odd bet:")
                        continue
                except: 
                    print("please type one of the following pass, don't pass or odd bet:")
                    continue
                else:
                    break
                           
        else:
            print("please type y or n as an input") #only allows y or n answer to go further
            Bets.betting_turn(self) #send the user back to the beginning of the function
    
    def ingest_bet(self): #ingest bet only allows positive integer number unless it does not exceed bank_roll
                
        while True:
            self.bet_amount = input("bet_amount:")
            try:         
                if int(self.bet_amount)<0:
                    print("negative vibe")
                    continue
                elif int(self.bet_amount)==0:
                    print("You can not have 0 bet amount")
                    continue 
                else:
                    if int(self.bet_amount) > int(self.bankroll):
                        Bets.Insufficient_funds(self)
                    else:
                        value2 = int(self.bet_amount)
                        print("you good boi")
            except:
                try:
                    value2 = float(self.bet_amount)
                    print("Please enter the value as an integer")
                    continue
                except ValueError:
                    print("That's not a number!")
                    continue
            else: 
                self.bet_amount = int(self.bet_amount) #just to be sure that bet_amount is saved as an integer value
                break      
    
    def ingest_bet2(self): #this is copied for odd_bets. When there is two active bets in the game, it is not possible to hold two different values with same self name.
                
        while True:
            self.bet_amount2 = input("bet_amount:")
            try:         
                if int(self.bet_amount2)<0:
                    print("Negative vibe")
                    continue
                elif int(self.bet_amount2)==0:
                    print("You can not have 0 bet amount")
                    continue 
                else:
                    if int(self.bet_amount2) > int(self.bankroll):
                        Bets.Insufficient_funds(self)
                    else:
                        value2 = int(self.bet_amount2)
                        print("you good boi")
            except:
                try:
                    value2 = float(self.bet_amount)
                    print("Please enter the value as an integer")
                    continue
                except ValueError:
                    print("That's not a number!")
                    continue
            else: 
                self.bet_amount2 = int(self.bet_amount2)
                break           
        
    def bet_won(self): #updating bankroll after win, stating bet_type and telling bankroll left
        self.bankroll += 2*self.bet_amount
        print(f"You won and earned {self.bet_amount}$ on this turn. Remaining balance is {self.bankroll}.")
            
    def bet_lost(self): #stating losing bet_type and amount and updates bankroll left
        print(f"You lost {self.bet_amount}$ on this turn. Remaining balance is {self.bankroll}.")
    
    def bet_made(self):
        if self.bet_amount != 0:
            print("bet is made")
    
    def shooter(self): #when sum of the dices are different than 2,3,7,11; this phase starts
        if self.point_on != 0:
            oddicin = input("Do you want to bet on odds? (y/n)") #it is not necessary but it allows to bet on numbers such 4,5,6,8,9 and 10.
            if oddicin == "y": #it allows betting now
                Bets.odds(self)#this is for betting on odds
                if self.total == self.point_on:
                    if self.pas == 1: #these numbers were assigned earlier to identify pass_line and do_not_pass line
                        Bets.bet_winner(self) 
                    elif self.pas == 2:
                        Bets.bet_loser(self)
                elif self.total == 7:
                    if self.pas == 1:
                         Bets.bet_loser(self) 
                    elif self.pas ==2:
                         Bets.bet_winner(self)
                else:
                    Bets.shooter(self) #if the conditions above is not satified, go back to the beginning of the function
            else: #if the user do not want to bet on odds
                Dice.roll(self)
                if self.total == self.point_on:
                    if self.pas == 1:
                        Bets.bet_winner(self) 
                    elif self.pas == 2:
                        Bets.bet_loser(self)
                elif self.total == 7:
                    if self.pas == 1:
                        Bets.bet_loser(self) 
                    elif self.pas ==2:
                        Bets.bet_winner(self)
                else:
                    Bets.shooter(self)  #if the conditions above is not satified, go back to the beginning of the function
        else:
            print("This round is over after the first roll")
    
    def bet_loser(self):
        print(f"{self.bet_type.upper()} was the current bet.")
        Bets.bet_lost(self)
    def bet_winner(self):
        print(f"{self.bet_type.upper()} was the current bet.")
        Bets.bet_won(self)
                
    def odds(self): #odd function is built accordingly 
        choose_number = int(input("Please enter one of the following: 4,5,6,8,9,10:"))
        Bets.ingest_bet2(self)
        value3 = self.bet_amount*3 #to check maximum allowable bets for each dice sum
        value4 = self.bet_amount*4
        value5 = self.bet_amount*5
        if choose_number == 6 or choose_number == 8: #so each condition allowable sum is different.
            if self.bet_amount2 > value3: #if it exceeds certain values, it recalls ingest_betting function.
                print(f"maximum allowable bet is {value3}")
                Bets.ingest_bet2(self)
            else:
                pass
        elif choose_number == 5 or choose_number == 9:
            if self.bet_amount2 > value4:
                print(f"maximum allowable bet is {value4}")
                Bets.ingest_bet2(self)
            else:
                pass
        elif choose_number == 4 or choose_number == 10:
            if self.bet_amount2 > value5:
                print(f"maximum allowable bet is {value5}")
                Bets.ingest_bet2(self)
            else:
                pass
        else:
            print("Type only following numbers: 4,5,6,8,9,10:") #if the input is other than those numbers, asks for input re-entry.
            Bets.odds(self)
        self.bankroll -= self.bet_amount2 
        Dice.roll(self)
        if choose_number == self.total:
            self.bankroll += 2*self.bet_amount2
        else:
            print(f"you lost {self.bet_amount2}$ bet on coming odds of {choose_number}.")
            pass
    def refresh(self): #this is for refreshing. Created for error handling purposes.
        self.total = 0
        self.dice1 = 0
        self.dice2 = 0
        self.bet_amount = 0
        self.bet_type = "None"
        self.point_on = 0
```
